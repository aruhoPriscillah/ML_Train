```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('IDH+test.data_Dryad.csv')
```

```python
# In this dataset, the 'Comment' column contains no information.
df_cleaned = df.drop(columns=['Comment'])
```

```python
# Some columns use '0' to mean 'no'. We also want to standardize text to lowercase.
def clean_categorical_logic(val):
    if pd.isna(val):
        return 'Unknown'
    val = str(val).strip().lower()
    return 'no' if val == '0' else val

# For malaria test results (RTDMP, BSMMP), '0' usually means 'negative'
def clean_malaria_test(val):
    if pd.isna(val):
        return 'Unknown'
    val = str(val).strip().lower()
    return 'negative' if val == '0' else val
    # Apply cleaning logic to relevant columns
df_cleaned['Compliant'] = df_cleaned['Compliant'].apply(clean_categorical_logic)
df_cleaned['ART'] = df_cleaned['ART'].apply(clean_categorical_logic)
df_cleaned['Septrin'] = df_cleaned['Septrin'].apply(clean_categorical_logic).replace('m', 'unknown')

df_cleaned['RTDMP'] = df_cleaned['RTDMP'].apply(clean_malaria_test)
df_cleaned['BSMMP'] = df_cleaned['BSMMP'].apply(clean_malaria_test)
```

```python
# Fill missing complaint descriptions with 'None'
df_cleaned['Typeofcomplaint'] = df_cleaned['Typeofcomplaint'].fillna('None')

# Handle missing numerical counts (e.g., frequency of net usage)
# If missing, we assume 0 usage in this context
df_cleaned['FreqofLLITN2'] = df_cleaned['FreqofLLITN2'].fillna(0)

# ARTstartyear: If missing, set to 0 (or you could use the median if only treating active patients)
df_cleaned['ARTstartyear'] = df_cleaned['ARTstartyear'].fillna(0)
```

```python
df_cleaned.to_csv('cleaned_dataset.csv', index=False)

print("Cleaning Complete!")
print(f"Original Shape: {df.shape}")
print(f"Cleaned Shape: {df_cleaned.shape}")
print("\nFirst 5 rows of cleaned data:")
display(df_cleaned.head())
```
